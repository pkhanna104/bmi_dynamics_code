```python
from sklearn.linear_model import Ridge

#### Randomly generate X data
Npts = 200
X = np.random.randn(Npts, 2)

#### Construct an A matrix (decay only) and compute X_{t+1}
A = np.array([[.9, 0.], [0, .9]])
Xt = np.dot(A, X.T).T

### Plot "flow"
f, ax = plt.subplots()
keep = []
for i in range(Npts):
    if X[i, 1] > 0:
        ax.plot([X[i, 0], Xt[i, 0]], [X[i, 1], Xt[i, 1]], 'k-')
        keep.append(i)
    else:
        ax.plot([X[i, 0], Xt[i, 0]], [X[i, 1], Xt[i, 1]], 'b-')
        
### Post hoc fit these points 
keep = np.array(keep)

### Try fitting data; 
X_train = X[keep, :]
Y_train = Xt[keep, :]

model = Ridge(alpha=1000, fit_intercept=True)
model.fit(X_train, Y_train)

#### Plot the estimated intercept; 
ax.plot(model.intercept_[0], model.intercept_[1], 'r*')
print('A, dyn matrix ', model.coef_)
```

```python
for i_a, animal in enumerate(['grom', 'jeev']): 
    f, ax = plt.subplots()
    data = pickle.load(open(analysis_config.config[animal+'_pref'] + 'LDS_maxL_ndims.pkl', 'rb'))
    for i_d in range(analysis_config.data_params['%s_ndays'%animal]):
        
        mn = np.mean(data[i_d, 'LLs'][0, :, :], axis=0)
        mnx = mn - mn[0]
        
        sem = np.std(data[i_d, 'LLs'][0, :, :] - mn[0], axis=0)/np.sqrt(5)
        ax.errorbar(data[i_d, 'dims']+0.1*np.random.randn(), mnx, sem)
    ax.set_xlabel('Dimensions -- state space')
    ax.set_ylabel('Log likelihood')
    ax.set_title('Animal %s'%animal)
```

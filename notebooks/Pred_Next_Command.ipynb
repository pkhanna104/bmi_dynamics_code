```python
import analysis_config
from online_analysis import generalization_plots, generate_models, util_fcns
from online_analysis.generalization_plots import DataExtract
```

```python
f, ax = plt.subplots()
f1, ax1 = plt.subplots()

for i_a, animal in enumerate(['jeev']):#, 'jeev']):
    for day_ix in range(4):#analysis_config.data_params['%s_ndays'%animal]):
        
        #### Get true data ####
        dataObj = DataExtract(animal, day_ix, model_nm = 'hist_1pos_0psh_2spksm_1_spksp_0', 
                model_set_number = 6, nshuffs=0)
        dataObj.load()

        if animal == 'grom':
            KG, KG_null_proj, KG_potent_orth = generate_models.get_KG_decoder_grom(day_ix)
        elif animal == 'jeev':
            KG, KG_null_proj, KG_potent_orth = generate_models.get_KG_decoder_jeev(day_ix)
            
        null, pot = generate_models.decompose_null_pot(0.1*dataObj.spks, dataObj.push, KG, KG_null_proj, KG_potent_orth)
        null_tm1, pot_tm1 = generate_models.decompose_null_pot(0.1*dataObj.spks_tm1, dataObj.push_tm1, KG, 
                                                               KG_null_proj, KG_potent_orth)

        ## Dynamics R2 of null / pot / combined ####
        T = dataObj.spks.shape[0]
        spks_pred, _, _ = generalization_plots.train_and_pred(0.1*dataObj.spks_tm1, 0.1*dataObj.spks, 
                                                              dataObj.push[:, [3, 5]], 
                                                              np.arange(T), np.arange(T), 0, KG, add_mean = None, 
                                                              skip_cond = True)
        spks_r2 = util_fcns.get_R2(0.1*dataObj.spks, spks_pred)
        KG = np.array(KG)
        act_r2 = util_fcns.get_R2(dataObj.push[:, [3, 5]], np.dot(KG, spks_pred.T).T)
        ax1.plot(day_ix + 10*i_a, act_r2, 'k*')
        
        #### NULL ####
        null_pred, _, _ = generalization_plots.train_and_pred(null_tm1, null, dataObj.push[:, [3, 5]], 
                                                        np.arange(T), np.arange(T), 0, KG, add_mean = None, 
                                                        skip_cond = True)
        null_r2 = util_fcns.get_R2(null, null_pred)

        #### POT ####

        pot_pred, _, _ = generalization_plots.train_and_pred(pot_tm1, pot, dataObj.push[:, [3, 5]], 
                                                        np.arange(T), np.arange(T), 0, KG, add_mean = None, 
                                                        skip_cond = True)
        pot_r2 = util_fcns.get_R2(pot, pot_pred)

        ax.plot(day_ix + 10*i_a, spks_r2, 'k*')
        ax.plot(day_ix + 10*i_a, null_r2, 'g*')
        ax.plot(day_ix + 10*i_a, pot_r2, 'r*')
        print('day %d: spks r2 %.3f, null r2 %.3f, pot r2 %.3f' %(day_ix, spks_r2, null_r2, pot_r2))
        
        shuff = []; shuff_act = []; 
        #### Now do some rolling; 
        for i in range(10): 
            
            #### Roll this guy ; 
            ix = np.arange(T)
            roll_ix = np.roll(ix, np.random.randint(0, T))
            
            spks_0 = null[roll_ix] + pot; 
            spks_1 = null_tm1[roll_ix] + pot_tm1; 
            
            spks_pred_i, _, _ = generalization_plots.train_and_pred(spks_1, spks_0, 
                                                              dataObj.push[:, [3, 5]], 
                                                              np.arange(T), np.arange(T), 0, KG, add_mean = None, 
                                                              skip_cond = True)
            r2i = util_fcns.get_R2(spks_0, spks_pred_i)
            shuff.append(r2i)
            KG = np.array(KG)
            r2i2 = util_fcns.get_R2(np.dot(KG, spks_0.T).T, np.dot(KG, spks_pred_i.T).T)
            shuff_act.append(r2i2)
        
        util_fcns.draw_plot(10*i_a + day_ix, shuff, 'k', np.array([1.,1.,1.,0.]), ax)
        util_fcns.draw_plot(10*i_a + day_ix, shuff_act, 'k', np.array([1.,1.,1.,0.]), ax1)
        
ax.set_xlim([-1, 5])
ax1.set_xlim([-1, 5])
```

```python
ndays = dict(grom=9, jeev=4)
for i_a, animal in enumerate(['grom', 'jeev']):
    su_pooled = []
    pop_pooled = []
    
    for day_ix in range(ndays[animal]):
        day_mn = np.mean(x[animal, day_ix]['su', 'true'])
        
        KG = util_fcns.get_decoder(animal, day_ix)
        nneurons = float(KG.shape[1])
        
        print('Nneur %d' %nneurons)
        shuff = np.vstack((x[animal, day_ix]['su', 'shuff'])) #ncommands x 1000
        assert(shuff.shape[1] == 1000)
        shuff_mn = np.mean(shuff, axis=0)

        ix = np.nonzero(shuff_mn <= day_mn)[0]
        print('Animal %s, Day_ix %d, pv = %.5f, day_mn %.3f, shuff_mn %.3f, shuff 5th %.3f'%(animal, 
                                                                                             day_ix, 
                                                                                             float(len(ix))/1000.,
                                                                                            day_mn, np.mean(shuff_mn),
                                                                                            np.percentile(shuff_mn, 5)))
        su_pooled.append([day_mn, shuff_mn])
        
        pop_day_mn = np.mean(np.array(x[animal, day_ix]['pop', 'true'])/nneurons)
        shuff_pop = np.mean(np.vstack((x[animal, day_ix]['pop', 'shuff']))/nneurons, axis=0)
        
        ix = np.nonzero(shuff_pop <= pop_day_mn)[0]
        print('POP: Animal %s, Day_ix %d, pv = %.5f, day_mn %.3f, shuff_mn %.3f, shuff 5th %.3f'%(animal,
                                                                                                  day_ix,
                                                                                                  float(len(ix))/1000.,
                                                                                                  pop_day_mn, 
                                                                                                  np.mean(shuff_pop),
                                                                                                  np.percentile(shuff_pop, 5)))
        pop_pooled.append([pop_day_mn, shuff_pop])
        
    #### pooled: 
    print('############')
    print('POOLED')
    
    mn = np.mean([d[0] for d in su_pooled])
    shf = np.mean(np.vstack(([d[1] for d in su_pooled])), axis=0)
    ix = np.nonzero(shf <= mn)[0]
    
    print('Animal %s,POOLED, pv = %.5f, day_mn %.3f, shuff_mn %.3f, shuff 5th %.3f'%(animal, 
                                                                                         float(len(ix))/1000.,
                                                                                        mn, np.mean(shf),
                                                                                        np.percentile(shf, 5)))

    
    pop_mn = np.mean([d[0] for d in pop_pooled])
    pop_shf = np.mean(np.vstack(([d[1] for d in pop_pooled])), axis=0)
    pop_ix = np.nonzero(shf <= mn)[0]
    
    print('Animal %s,POP POOLED, pv = %.5f, day_mn %.3f, shuff_mn %.3f, shuff 5th %.3f'%(animal, 
                                                                                         float(len(pop_ix))/1000.,
                                                                                        pop_mn, np.mean(pop_shf),
                                                                                        np.percentile(pop_shf, 5)))
```

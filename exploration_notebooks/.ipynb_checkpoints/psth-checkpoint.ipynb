```python
#Map between Task Code (tc) and Task {'co', 'obs'}
task2tc = {'co':data['task_entries'][0], 'obs':data['task_entries'][1]}
tc_list = [item for sublist in data['task_entries'] for item in sublist]
tc2task = {}
for k,v_list in task2tc.items():
    for v in v_list:
        tc2task[v] = k
print(tc_list)
print(tc2task)
```

```python
num_targets = 8
tc_meta = {}
tc_trials = {}
task_list = ['co', 'obs']
task2data_list = {}
task2data_list['co'] = ['cursor_pos', 'cursor_state', 'neural_push',  'binned_spk_cnts']
task2data_list['obs'] = ['cursor_pos', 'cursor_state', 'neural_push',  'binned_spk_cnts', 'obstacle_size', 'obstacle_location']
target_color = ['maroon', 'orangered', 'darkgoldenrod', 'olivedrab', 'teal', 'steelblue', 'midnightblue', 'darkmagenta']
sample_period = 0.1 #seconds
target_hold_time = 0.2 #seconds

for tc in tc_list: 
    task = tc2task[tc]
    tc_meta[tc] = {}
    tc_trials[tc] = []
    
    num_trials = int(data[tc, 'trial_ix'].max()+1)
    ttype = np.array([data[tc, 'target_index'][np.where(data[tc, 'trial_ix'] == i_trial)[0][0]] for i_trial in range(num_trials)]).astype(int)
    target_pos = np.array([data[tc, 'target_pos'][np.where(data[tc, 'target_index'] == i_target)[0][0], :] for i_target in range(num_targets)])
    num_trials_per_target = [np.sum(ttype==i) for i in range(num_targets)]
    target_trial_idxs = {}
    for i in range(num_targets):
        target_trial_idxs[i] = np.where(ttype==i)[0]
    if task == 'obs':
        #obs_pos: one obs position per target
        obs_pos = np.array([data[tc, 'obstacle_location'][target_trial_idxs[i][0], :] for i in range(num_targets)])
        #obs_size_per_target: for each target, an array of obstacle size
        obs_size_per_target_trial = [data[tc, 'obstacle_size'][target_trial_idxs[i],:] for i in range(num_targets)]
    #for each target, what are the obstacle sizes
    
    #Loop trials: 
    for i in range(num_trials):
        trial_data = {}
        for d in task2data_list[tc2task[tc]]:
            trial_data[d] = np.array(data[tc, d][i])
        #time of trial: 
        trial_data['len'] = sample_period*trial_data['cursor_pos'].shape[0]
        tc_trials[tc].append(trial_data)
    
    #ASSIGN:
    tc_meta[tc]['task'] = tc2task[tc]
    tc_meta[tc]['ttype'] = ttype
    tc_meta[tc]['target_pos'] = target_pos
    if task == 'obs':
        tc_meta[tc]['obs_pos'] = obs_pos
        tc_meta[tc]['obs_size_per_target_trial'] = obs_size_per_target_trial
    tc_meta[tc]['num_trials'] = num_trials
    tc_meta[tc]['num_trials_per_target'] = num_trials_per_target
    tc_meta[tc]['target_trial_idxs'] = target_trial_idxs
    tc_meta[tc]['sample_period'] = sample_period
    tc_meta[tc]['target_hold'] = target_hold_time
    
```

```python
#Variables to initialize psth matrix: 


#NUM NEURONS
test_tc = task2tc['co'][0]
test_trial = tc_trials[test_tc][1] 
num_neurons = test_trial['binned_spk_cnts'].shape[1]
print('num neurons', num_neurons)

#DATA DIMENSIONS
d_list = ['cursor_state', 'neural_push', 'binned_spk_cnts']
d_dic = {('cursor_state', 'dim'):4, ('neural_push', 'dim'):4, ('binned_spk_cnts', 'dim'):num_neurons}

#NUM TRIALS
num_trials_dic = {}
for task, tc_list in task2tc.items():
    for target in range(num_targets):
        num_trials_dic[task, target] = 0
        for tc in tc_list: 
            num_trials_dic[task, target] += len(tc_meta[tc]['target_trial_idxs'][target])
```

```python
max_trial_len = 80
accum = {} #task, target, data
for task, tc_list in task2tc.items():
    for d in d_list:
        accum[task, 'all', d] = []
    for target in range(num_targets):
        num_trials = num_trials_dic[task, target]
        #Initialize
        for d in d_list:
            num_var = d_dic[(d, 'dim')]
            accum[task, target, d] = np.ma.array(np.zeros((max_trial_len, num_var, num_trials)), 
                                              mask=np.ones((max_trial_len, num_var, num_trials)))
        trial_cnt = -1
        for tc in tc_list:
            print(task, target, tc)
            trial_list = tc_meta[tc]['target_trial_idxs'][target]
            for trial in trial_list:
                trial_cnt += 1
                data = tc_trials[tc][trial]
                for d in d_list: 
                    d_i = data[d]
                    n_i = min(d_i.shape[0], max_trial_len)
                    accum[task, target, d].data[0:n_i,:,trial_cnt] = d_i[0:n_i,:]
                    accum[task, target, d].mask[0:n_i,:,trial_cnt] = 0
                    
```

```python
#PSTH Result: 
psth = {}
for task in task_list: 
    for target in range(num_targets):
        for d in d_list: 
            data = accum[task, target, d]
            psth[task, target, d, 'mean'] = data.mean(axis=2)
            psth[task, target, d, 'sem'] = data.std(axis=2)/np.sqrt(data.count(axis=2))
```

```python
#ACCUM POSITION AND POSITION DELTA: 

#1) Accumulate + Hist position data: 
#2) Accumulate p, pd 

accum = {}
hist_dic = {}
prefix = 9

#BINs: 
#------------------------------------------------------------------------
delta = 1.5
lim = np.array([-15.0, 15.0])
lim_dic = {0:lim, 1:lim}
g = np.arange(lim[0], lim[1]+1, delta)
e = np.vstack((g[:-1], g[1:]))
num_bins = e.shape[1]
bin_dic = {0:e, 1:e}

#ACCUM
#------------------------------------------------------------------------
d_list = ['p', 'pd']
for task, tc_list in task2tc.items():
    for d in d_list:
        accum[task, 'all', d] = []
    for target in range(num_targets):
        for d in d_list: 
            accum[task, target, d] = []
        for tc in tc_list:
            print(task, target, tc)
            trial_list = tc_meta[tc]['target_trial_idxs'][target]
            for trial in trial_list:
                pos = tc_trials[tc][trial]['cursor_pos'][prefix:,:]
                p = pos[:-1,:]
                pd = pos[1:,:]-pos[:-1,:]
                #APPEND:
                accum[task, target, 'p'].append(p)
                accum[task, target, 'pd'].append(pd)
        for d in d_list:
            accum[task, target, d] = np.array(np.vstack(accum[task, target, d]))
        #position out of bounds:
        accum[task, target, 'p'], del_idxs, out_bound_over_dim = bmi_b.remove_data_out_bound(accum[task, target, 'p'], lim_dic)
        for d in d_list:
            if d != 'p':
                accum[task, target, d] = np.delete(accum[task, target, d], del_idxs, axis=0)
        #Pool across: 
        for d in d_list: 
            accum[task, 'all', d].append(accum[task, target, d])
    for d in d_list:

        accum[task, 'all', d] = np.array(np.vstack(accum[task, 'all', d]))
```

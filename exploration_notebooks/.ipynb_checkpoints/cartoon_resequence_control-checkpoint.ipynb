### Plot decaying dynamics

```python
#Plt exp decay
rot_mat = bmi_b.rot2D_mat(45*np.pi/180)

n = np.arange(0,40)

alpha_y = 0.9
alpha_x = 0.5
y0=(1-0.25*0)
x0=(1+0.25*0)
y = y0*alpha_y**(n)
x = x0*alpha_x**(n)
yn = -y
xn = -x

proj = np.dot(rot_mat, np.array([y,x]))


plt.figure()
r = plt.plot(x,y, '.-')
r = plt.plot(xn,yn, '.-')
plt.plot(np.linspace(-0.5,0.5,2), np.linspace(0.5,-0.5,2))
plt.plot(np.linspace(-0.5,0.5,2), np.linspace(-0.5,0.5,2))


# r = plt.axis('square')
# r = plt.axis('equal')
plt.xlabel('x')
plt.ylabel('y')
plt.title('y vs x')
r = plt.axis('equal')
# plt.xlim((-1.1,1.1))
# plt.ylim((-1.1,1.1))

plt.xlabel('time')
plt.legend()
plt.title('time vs x, x')

plt.figure()
plt.plot(n, x, label='x')
plt.plot(n, y, label='y')
plt.plot(n, proj[0,:], label='proj on y+x=0')
# plt.plot(n, proj[1,:], label='proj on y-x=0')

plt.xlabel('time')
plt.legend()
plt.title('time vs x, x')
# plt.axis('square')
```

```python
#Loop 
#dynamics: 
A = np.array([[0.9, 0], [0, 0.5]])

#initial conditions: 
# init = np.array([[1,1], [-1,1], [1,-1], [-1,-1]])
starter = np.array([[-1,1], [1,-1]])

init = starter*0.5
init = np.vstack((init, 1*starter))
# init = np.vstack((init, 0.5*init))

traj_dic = {}
N = 40
for i, x0 in enumerate(init): 
    traj_dic[i,'x0'] = x0
    traj_dic[i] = np.zeros((2,N))
    traj_dic[i][:,0] = x0.T
    for n in np.arange(1,N): 
        traj_dic[i][:,n] = np.dot(np.linalg.matrix_power(A,n), x0.T)
        
```

```python
dyn_dic = {}
dyn_dic['decay'] = traj_dic
```

```python
plt.figure()
plt.plot(np.linspace(-1.0,1.0,2), np.linspace(1.0,-1.0,2), '--', color='k')
plt.plot(np.linspace(-1.0,1.0,2), np.linspace(-1.0,1.0,2), color='k')

for i in np.arange(0, init.shape[0]):
    plt.plot(traj_dic[i][0,:], traj_dic[i][1,:], '-')


plt.axis('equal')
```

### Plot flow field

```python
#Plot dynamics flow field
#Inefficient strategy:
#At each grid point, evaluate the delta.  Quiver it.  

#Dynamics: 
save_bool = True

A = np.array([[0.9, 0], [0, 0.5]])
I = np.eye(2)

num_div = 6
grid_lim = 1
x_grid = np.hstack((np.linspace(-grid_lim,0,num_div), np.linspace(0,grid_lim,num_div)[1:]))
y_grid = x_grid

gain = 1
width = 0.002
plt.figure()
for x in x_grid:
    for y in y_grid:
        d = np.dot((A-I), np.array([x,y]).T)
        plt.quiver(x, y, d[0], d[1], angles='xy', scale=1/gain, scale_units='xy', width=width, color='k') #, zorder=10
plt.axis('square')

plt.plot(np.linspace(-grid_lim,grid_lim,2), np.linspace(grid_lim,-grid_lim,2), '--', color='k')
plt.plot(np.linspace(-grid_lim,grid_lim,2), np.linspace(-grid_lim,grid_lim,2), color='k')

for i in np.arange(0, init.shape[0]):
    plt.plot(traj_dic[i][0,:], traj_dic[i][1,:], '-', linewidth=3)
    
print(A)
```

```python
A = reg.coef_
print(A.shape)
# b = reg.intercept_
# print(b.shape)
```

```python
w,v = np.linalg.eig(A)
```

```python
#sanity check:
w_abs_mat = np.abs(np.diag(w)) 
test = np.abs(np.dot(np.linalg.inv(v), np.dot(A, v)))

plt.figure()
plt.imshow(w_abs_mat)

plt.figure()
plt.imshow(test)
```

```python
#sanity check: 
rhs = np.abs(np.dot(A,v))
lhs = np.abs(np.dot(v,np.diag(w)))

plt.figure()
plt.imshow(lhs)

plt.figure()
plt.imshow(rhs)
```

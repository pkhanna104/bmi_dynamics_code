```python
#### Plot R2 of models against one another; 
models = dict()
models[6] = 'hist_1pos_0psh_0spksm_1_spksp_0'
models[11] = 'hist_1pos_0psh_0spksm_1_spksp_0_latentLDS'
ndays = dict(grom=9, jeev=4)
fitintc = dict(Ridge = [True, False], LatentLDS = [False])
suffx = dict(Ridge = [''], LatentLDS = ['', '_ndim15'])

#### Iterate through 
for animal in ['jeev', 'grom']: 
    f, ax = plt.subplots(figsize = (5, 7))
    mns = np.zeros((4, ndays[animal]))
    xlab = []
    cnt = -1; 
    for _, (model_set, modnm) in enumerate(zip([6, 11], ['Ridge', 'LatentLDS'])): 
        p = analysis_config.config[animal+'_pref'] + 'tuning_models_'+animal+'_model_set%d_task_spec_pls_gen_match_tsk_N'%(model_set)
        
        for fit_intc in fitintc[modnm]: 
            for ndim in suffx[modnm]:
                
                skip = False
                
                if fit_intc and model_set == 6:
                    dat = pickle.load(open(p + '.pkl', 'rb'))
                
                elif fit_intc == False and model_set == 6:
                    dat = pickle.load(open(p + '_no_intc.pkl', 'rb'))
                
                elif model_set == 11 and ndim == '':
                    dat = pickle.load(open(p + '_no_intc.pkl', 'rb'))
                
                elif model_set == 11 and ndim == '_ndim15':
                    dat = pickle.load(open(p + '_no_intc_ndim15.pkl', 'rb'))
                    
                else:
                    print('Skipping model set %d, fit intc %s, ndim %s' %(model_set, str(fit_intc), ndim))
                    skip = True

                if skip:
                    pass
                else:
                    cnt += 1
                    xlab.append('%s\nfit intc %s, ndim %s' %(modnm, fit_intc, ndim))
                    
                    #### Plot bars using this data; 
                    for i_d in range(ndays[animal]):

                        ### General dynamics ####
                        spks_true = dat[i_d, 'spks']
                        spks_pred = dat[i_d, models[model_set]][:, :, 2]
                        assert(spks_true.shape == spks_pred.shape)
                        r2 = util_fcns.get_R2(spks_true, spks_pred, pop = True)
                        mns[cnt, i_d] = r2
    ### Plot means; 
    for i in range(4):
        ax.bar(i, np.mean(mns[i, :]), width=.9)
        ax.errorbar(i, np.mean(mns[i, :]), np.std(mns[i, :])/np.sqrt(mns.shape[1]), marker='|')
    for i_d in range(ndays[animal]):
        ax.plot(range(4), mns[:, i_d], '-', color='gray')
    ax.set_title(animal)
    ax.set_xticks(np.arange(4))
    ax.set_xticklabels(xlab, rotation=45, fontsize = 14)
    f.tight_layout()
```

```python
### Example; 
A = np.mat([[np.cos(np.pi/4.), -np.sin(np.pi/4.)], [np.sin(np.pi/4.), np.cos(np.pi/4.)]])
B = np.mat([[0.], [1.,]])
Q = np.mat(np.eye(2))
R = .500000

ctl = feedback_controllers.LQRController(A, B, Q, R)
K = ctl.dlqr(A, B, Q, R)

Acl = np.squeeze(np.array(A - np.dot(B, K)))
```

```python
_, ev = np.linalg.eig(A)
_, evx = np.linalg.eig(Acl)
```

```python
### Plot A matrix flow field; 
f, ax = plt.subplots(ncols = 2, figsize = (10, 5))
for axi in ax:
    axi.axis('equal')
plot_flow_field_utils.plot_flow(np.squeeze(np.array(A)), ax[0], cmax = 5., width = .05, scale=1.)
ax[0].set_title('A matrix')
plot_flow_field_utils.plot_flow(np.squeeze(np.array(Acl)), ax[1], cmax = 5., width = .05, scale=1.)
ax[1].set_title('A - BK matrix, R = %.1f, Q = %.1f' %(R, Q[0, 0]))
```

```python
X = []
for x1 in np.arange(-5,5,.1):
    for x2 in np.arange(-5, 5, .1):
        x = np.array([[x1], [x2]])
        u = -1*np.dot(K, x)
        X.append(np.hstack(( [[x1], [x2], np.squeeze(np.array(u))])) )
X = np.vstack((X))
f, ax = plt.subplots(figsize = (10, 5), ncols = 2)
cax =ax[0].scatter(X[:, 0], X[:, 1], s=3, c=X[:, 2], cmap='viridis')
cax =ax[1].scatter(X[:, 0], X[:, 1], s=3, c=X[:, 3], cmap='viridis')
```
